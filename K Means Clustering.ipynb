```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
#Artificial created cluster lebels to check the predicted lebels are correct or not
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
lm = KMeans(n_clusters=4)
```

```python
lm.fit(data[0])
```

```python
lm.cluster_centers_
```

```python
#predicted lebel of clusters
lm.labels_
```

```python
fig ,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(15,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=lm.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
lm = KMeans(n_clusters=2)
```

```python
lm.fit(data.drop(['Unnamed: 0','Private'],axis=1))
```

```python
lm.cluster_centers_
```

```python
def convertor(cluster):
    if cluster == 'Yes':
        return 1
    else:
        return 0
```

```python
data['cluster'] = data['Private'].apply(convertor)
```

```python
data['cluster']
```

```python
lm.labels_
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(data['cluster'],lm.labels_))
print('\n')
print(classification_report(data['cluster'],lm.labels_))
```
